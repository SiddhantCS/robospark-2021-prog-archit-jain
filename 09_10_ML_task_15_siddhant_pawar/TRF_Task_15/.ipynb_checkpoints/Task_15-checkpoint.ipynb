```python
# Making a list of missing value types
missing_values = ["NaN"]
df = pd.read_csv("data.csv", na_values = missing_values)
```

```python
# Total missing values for each feature
print(df.isnull().sum())
```

```python
# Total number of missing values
print(df.isnull().sum().sum())
```

```python
 #Clean the data
```

```python
# Replace using median 
median = df['age'].median()
df['age'].fillna(median, inplace=True)
```

```python
median = df['fare'].median()
df['fare'].fillna(median, inplace=True)
```

```python
df
```

```python
# the columns which provide no info of relevance are dropped from the dataframe 
# axis - 0 : rows , 
df.drop(labels = {'name','passenger_id' ,'home.dest', 'pclass' ,'parch','cabin','boat','body','ticket','embarked'} , axis = 1 , inplace = True)
```

```python
df
```

```python
#Cleaning of data done
```

```python
# Normalize any required columns
```

```python
#Using The maximum absolute scaling:
# copy the data
df_max_scaled = df.copy()
  
# apply normalization techniques on Column 1
column = 'age'
df_max_scaled[column] = df_max_scaled[column] /df_max_scaled[column].abs().max()
  
# view normalized data
display(df_max_scaled)
```

```python
#Using sklearn
from sklearn.preprocessing import MinMaxScaler
```

```python
# copy the data
df_sklearn = df_max_scaled.copy()
  
# apply normalization techniques
column = 'fare'
df_sklearn[column] = MinMaxScaler().fit_transform(np.array(df_sklearn[column]).reshape(-1,1))
  
# view normalized data  
display(df_sklearn)
```

```python
#Normalize any required columns 'Completed'
```

```python
#Encode the categorical data [ use label encoding only for simplicity ]
```

```python
df_copy = df_sklearn.copy()
```

```python
from sklearn import preprocessing
```

```python
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
 
# Encode labels in column 'Group'.
df_copy['sex']= label_encoder.fit_transform(df_copy['sex'])
 
df_copy['sex'].unique()
```

```python
# the sklearn package includes different regressors , classifiers and preprocessing functions 
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
#split dataset in features and target variable
feature_cols = ['sex', 'age', 'sibsp', 'fare']
X = df_copy[feature_cols] # Features
y = df_copy.survived # Target variable
```

```python
# for training the models we divide data into train and test set 
# a ratio in between 7:3 and 8:2 is followed [train:test]
# in our case we will take the ratio as 75:25

# X_train , X_test , y_train, y_test = model_selection.train_test_split(X , y ,test_size = 0.25 , shuffle = True , stratify=Y  )
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
# shuffle randomizes the order in which the data is divided
# stratify preserves the ratio of different classes in both train and test data 
```

```python
X_train.shape
```

```python
print(df_copy.isnull().sum())
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)

# Train Decision Tree Classifer
clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
print("Score:",metrics.accuracy_score(y_test, y_pred))
```
